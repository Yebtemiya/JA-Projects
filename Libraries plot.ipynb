```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 12, 8, 15, 7]
plt.xlabel('X-axis label')
plt.ylabel('Y-axis label')
plt.title('Line Plot')
plt.plot(x,y,linestyle='--', color='green', marker='o');
```

```python
import seaborn as sns
```

```python
categories = ['A', 'B', 'C', 'D']
values = [25, 40, 15, 30]

plt.bar(categories, values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot');
```

```python
data = [32, 45, 50, 28, 36, 42, 38, 25, 29, 52, 48, 60]
plt.title('Histogram Plot')
plt.xlabel('data')
plt.ylabel('frequency')
sns.histplot(data, bins=10, kde=True);
```

```python
x = [3, 5, 7, 10, 12, 15]
y = [25, 40, 30, 50, 35, 60]
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
sns.scatterplot(x=x, y=y);
```

```python
labels = ['Apples', 'Bananas', 'Oranges', 'Grapes']
sizes = [30, 25, 20, 15]
plt.pie(sizes, labels=labels, autopct='%.1f%%', startangle=0)
plt.axis('equal')
plt.title('Pie Chart');
```

```python
data = [45, 60, 75, 300, 80, 350, 90, 95, 100, 110, 120, 250, 200]
sns.boxplot(data, width=0.5)
plt.title('Box Plot');
```

```python
categories = ['A', 'B', 'C', 'D']
values = [25, 40, 15, 30]

sns.barplot(y=categories, x=values,orient='h')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Horizontal Bar Plot');
```
