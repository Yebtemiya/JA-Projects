```python
df=df.drop(['id','lat','long','country code','house_rules','license'], axis=1)
```

```python
df
```

```python
cat_columns = df.select_dtypes(include='object').columns
cat_modes = df[cat_columns].mode().iloc[0]
for column in cat_columns:
    df[column].fillna(cat_modes[column], inplace=True)
```

```python
for cols in df.columns:
    print(cols,df[cols].unique())
    print('')
```

```python
numerical_columns = df.select_dtypes(include='number').columns
means = df[numerical_columns].mean()
for column in numerical_columns:
    df[column].fillna(means[column], inplace=True)
```
